## Reading text file

```python
import json
f=open('march-2019.txt',encoding="utf8") #,

#creating an instance s which holds all the content of the text file( sample.txt)
s=f.read() 
```

```python
s=s.replace('\n',' ')
```

## To check garbage

```python
for i in range(795,896):
    st=s.find(str(i)+')')
    en=s.find(str(i+1)+')')
    #print(i)
    #--------to remove--Extra kachra---------#
    
    if 'GYANM GENERAL AWARENESS – OCTOBER 2019' in s[st:en]:
        print(i)
```

## Appending to list ques,ans

```python
li=[]
for i in range(795,896):
    st=s.find(str(i)+')')
    en=s.find(str(i+1)+')')
    print(i)
    #--------to remove--Extra kachra---------#
    
    #if 'GYANM GENERAL AWARENESS – OCTOBER 2019' in s[st:en]:
       # print(i)
    
    #----------X-----------------------#    
    string=s[st:en]    
    if ':' in string:
        x,y=string.split(':')
        li.append([x,y])
    elif '?' in string:
        a,b=string.split('?')
        li.append([a,b])
   # print(s[st:en],sep='\n\n')
```

```python
li
```

```python
o = open("march-2019.json", "w") 
json.dump(li, o, indent = 4, sort_keys = False) 
o.close() 
```
